```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("demographics.csv")
```

## Normallıq fərziyyəsinin numerikal metodla yoxlanılması

```python
#Shapiro-Wilk normallıq testi ilə (income) dəyişəninin normal paylanıb-paylanmadığını yoxlayacağıq.

statistic, p_value = stats.shapiro(data['income'])
```

```python
print("Shapiro-Wilk Test Results:")
print(f"Statistic: {statistic}")
print(f"p-value: {p_value}")
```

## Normallıq fərziyyəsinin qrafik  metodla yoxlanılması

```python
plt.hist(data['income'], bins=30)
plt.show()
```

```python
stats.probplot(data['income'], plot=plt)
plt.show()
```
