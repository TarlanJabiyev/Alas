## 1. Element-bə-element əməliyyatlar vs Matris Hasil

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])

print("A:\n", A)
print("B:\n", B)
```

```python
# Hadamard Vurması (element-elementə)
hadamard = A * B
print("Hadamard Hasil (A * B):\n", hadamard)

# Matris Hasil (Dot product)
dot_prod = A @ B  # və ya np.dot(A, B)
print("\nMatrisin Riyazi Hasili (A @ B):\n", dot_prod)
```

**Kommutativlik Testi:** $AB \neq BA$

```python
print("A @ B:\n", A @ B)
print("\nB @ A:\n", B @ A)
print("\nGördüyünüz kimi fərqli nəticələr alındı!")
```

```python
# Matrisin Riyazi Hasili (A @ B)
np.array([
    [A[0,0] * B[0,0], A[0,1] * B[0,1]],
    [A[1,0] * B[1,0], A[1,1] * B[1,1]]
])
```

```python
# Hadamard Hasil (A * B)
np.array([
    [A[0,0] * B[0,0] + A[0,1] * B[1,0], A[0,0] * B[0,1] + A[0,1] * B[1,1]],
    [A[1,0] * B[0,0] + A[1,1] * B[1,0], A[1,0] * B[0,1] + A[1,1] * B[1,1]]
])
```

## 3. Matrisin İzi (Trace) və Determinant

### 3.1 Matrisin İzi (Trace)
Kvadrat matrisin **izi** ($Tr(A)$), onun əsas diaqonalı üzərində yerləşən elementlərin cəminə bərabərdir.

**Əsas xassələr:**
1. **Xəttilik:** $Tr(A+B) = Tr(A) + Tr(B)$
2. **Siklik xassə:** $Tr(AB) = Tr(BA)$ (Baxmayaraq ki, $AB \neq BA$)

```python
# İzi (Trace) hesablamaq
M1 = np.array([[5, 2, 1], 
               [0, -3, 8], 
               [4, 1, 7]])

print("M1 matrisinin İzi:", np.trace(M1)) 
```

```python
M1[0,0] + M1[1,1] + M1[2,2]
```

```python
# Dövri xassənin yoxlanılması
M2 = np.array([[1, 0, 2], 
               [3, 4, 5], 
               [0, 2, 1]])

tr_AB = np.trace(M1 @ M2)
tr_BA = np.trace(M2 @ M1)

print(f"\nTr(M1 @ M2) = {tr_AB}")
print(f"Tr(M2 @ M1) = {tr_BA}")
print("Xassə doğrudurmu?", tr_AB == tr_BA)
```

### 3.2 Determinant və Onun Xassələri

Determinant fəzanın ölçü dəyişmə faktorudur. 

**Vacib Xassələr:**
1. $\det(AB) = \det(A) \cdot \det(B)$
2. $\det(kA) = k^n \det(A)$ (burada $n$ matrisin ölçüsüdür)
3. $\det(A^T) = \det(A)$

```python
# 1. 2x2 Determinant: det([[3,4],[2,5]]) = 15 - 8 = 7
A_2x2 = np.array([[3, 4], 
                  [2, 5]])
print("2x2 Determinant:", np.linalg.det(A_2x2))

# 2. 3x3 Determinant
B_3x3 = np.array([[1, 2, 3], 
                  [0, 1, 4], 
                  [5, 6, 0]])
print("3x3 Determinant:", round(np.linalg.det(B_3x3), 2))

# 3. Xassə yoxlanışı: det(2A) = 2^2 * det(A) = 4 * 7 = 28
k = 2
det_kA = np.linalg.det(k * A_2x2)
print(f"\ndet(2 * A_2x2) = {round(det_kA, 2)}")
```

## 4. Tətbiq: Cramer Qaydası

Aşağıdakı sistemi Cramer qaydası ilə həll edək:

$2x + 3y = 8$

$x - y = -1$

```python
C = np.array([[2, 3], 
              [1, -1]])
R = np.array([8, -1])

det_C = np.linalg.det(C)

det_C
```

```python
# x-i tapmaq üçün x sütununu dəyişdik
Cx = np.column_stack([R, C[:, 1]])
Cx
```

```python
det_Cx = np.linalg.det(Cx)
x = det_Cx / det_C

# y-i tapmaq üçün y sütununu dəyişdik
Cy = np.column_stack([C[:, 0], R])
det_Cy = np.linalg.det(Cy)
y = det_Cy / det_C

print(f"X = {round(x, 2)}, Y = {round(y, 2)}")
```
