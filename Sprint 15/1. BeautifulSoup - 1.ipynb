```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
soup = BeautifulSoup(page.text, 'html')
```

## HTML tags

```python
# soup.<tag>: Return content between opening and closing tag including tag

soup.title
```

```python
# soup.<tag>.string: Return string within given tag

soup.title.string
```

```python
# Find all the links within page’s <a> tags

soup.a
```

```python
all_a = soup.find_all('a')
print('Total number of URLs:',len(all_a)) 
all_a
```

```python
for link in all_a:
    print(link.get('href'))
```

## Table

```python
all_tables = soup.find_all('table') 
print('Total number of tables:',len(all_tables)) 
table = all_tables[0]
```

```python
world_titles = table.find_all('th')
world_titles
```

```python
len(world_titles)
```

```python
world_titles[1].text.strip()
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
world_table_titles
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
row_data = column_data[1:][0].find_all('td')
row_data
```

```python
row_data[1].text.strip()
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
