```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from plotnine.data import mpg
```

```python
data = mpg
data.head()
```

```python
sns.set_style("whitegrid") #dark, white, darkgrid, whitegrid, ticks
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='displ', y='cty', color='red')
plt.xlabel('mühərrikin həcmi')
plt.ylabel('millər')
plt.title('Yanacaq iqtisadiyyatı məlumatları\nScatterplot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='displ', y='cty', size='hwy', color='red')
plt.xlabel('mühərrikin həcmi')
plt.ylabel('millər')
plt.title('Yanacaq iqtisadiyyatı məlumatları\nScatterplot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='displ', y='cty', size='hwy', color='red')
plt.xlabel('mühərrikin həcmi')
plt.ylabel('millər')
plt.title('Yanacaq iqtisadiyyatı məlumatları\nScatterplot')
plt.yticks(range(8, 40, 2))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='displ', y='cty', hue='class', size='hwy', sizes=(20, 200), palette='viridis')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('City MPG')
plt.title('City MPG vs Engine Displacement by Vehicle Class')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='hwy', hue='drv', bins=30, multiple='stack')
plt.xlabel('Highway MPG')
plt.title('Highway Mileage Histogram by Transmission')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='class', hue='drv', palette='Dark2')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.title('Vehicle Count by Transmission and Drive Type')
plt.xticks(rotation=45)
plt.legend(title='Drive Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='hwy', x='class', palette='Set2')
plt.ylabel('Highway MPG')
plt.title('Highway Mileage by Class')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxenplot(data=data, x='year', y='hwy', hue='class', palette='Set3')
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.title('Highway MPG Distribution by Year and Vehicle Class')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
