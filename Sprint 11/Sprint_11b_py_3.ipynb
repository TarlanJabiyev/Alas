```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "S.No": range(1, 11),
    "qrup.1": [12, 23, 34, 45, 56, 67, 78, 89, 90, 1],
    "qrup.2": [123, 234, 345, 456, 567, 678, 789, 890, 901, 12],
    "qrup.3": [1234, 2345, 3456, 4567, 5678, 6789, 7890, 8901, 9012, 123]
})
df.head()
```

## pd.melt() funksiyası

```python
gather_df = pd.melt(df, id_vars=["S.No"], 
                    value_vars=["qrup.1", "qrup.2", "qrup.3"],
                    var_name="qrup", value_name="dəyər")
gather_df.head()
```

## str.split() funksiyası

```python
separate_df = gather_df.copy()
separate_df[["hissə", "nömrə"]] = separate_df["qrup"].str.split(".", expand=True)
separate_df.head()
```

## str.cat() funksiyası

```python
unite_df = separate_df.copy()
unite_df["qrup"] = unite_df[["hissə", "nömrə"]].agg(".".join, axis=1)
unite_df.head()
```

## pivot() funksiyası

```python
əvvəlki_hal = unite_df.pivot(index="S.No", columns="qrup", values="dəyər").reset_index()
əvvəlki_hal.head()
```
