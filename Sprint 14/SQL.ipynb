## SQL

```python
pd.read_sql_query("SELECT * FROM tweats", con)
```

```python
pd.read_sql_query("SELECT * FROM tweats WHERE date >= '2015-09-22'", con)
```

```python
pd.read_sql_query("SELECT * FROM comments LIMIT 5", con)
```

```python
pd.read_sql_query("""
    SELECT 
        SUM(tweat_id), 
        MAX(tweat_id), 
        MIN(tweat_id), 
        AVG(tweat_id) 
    FROM comments
""", con)
```

```python
pd.read_sql_query("""
    SELECT 
        SUM(tweat_id) AS sum_tweat_id, 
        MAX(tweat_id) AS max_tweat_id, 
        MIN(tweat_id) AS min_tweat_id, 
        AVG(tweat_id) AS avg_tweat_id 
    FROM comments
""", con)
```

```python
pd.read_sql_query("""
    SELECT 
        user_id,
        SUM(tweat_id) AS sum_tweat_id, 
        MAX(tweat_id) AS max_tweat_id, 
        MIN(tweat_id) AS min_tweat_id, 
        AVG(tweat_id) AS avg_tweat_id 
    FROM comments 
    GROUP BY user_id
""", con)
```

```python
pd.read_sql_query("SELECT * FROM comments ORDER BY user_id", con)
```

```python
pd.read_sql_query("SELECT * FROM comments ORDER BY user_id DESC", con)
```

```python
pd.read_sql_query("SELECT *, LENGTH(post) FROM tweats", con)
```

```python
pd.read_sql_query("SELECT COUNT(*) FROM tweats", con)
```

```python
pd.read_sql_query("SELECT DISTINCT date FROM tweats", con)
```

```python
pd.read_sql_query("SELECT DISTINCT * FROM comments", con)
```

```python
pd.read_sql_query("""
    SELECT 
        c.*, 
        t.post, 
        t.date 
    FROM comments c 
    LEFT JOIN tweats t 
    ON c.user_id = t.user_id
""", con)
```
