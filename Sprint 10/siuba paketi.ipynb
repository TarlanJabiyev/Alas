```python
import pandas as pd
from siuba import *
```

```python
url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
data = pd.read_csv(url)
data.head()
```

## arrange() funksiyasÄ±

```python
data >> arrange(_.lifeExp) >> _.head()
```

```python
data >> arrange(-_.lifeExp) >> _.head()
```

```python
data >> arrange(_.continent) >> _.head()
```

```python
data >> arrange(_.continent, -_.lifeExp) >> _.head()
```

## rename() funksiyasÄ±

```python
data >> rename(life_expectancy = _.lifeExp, gdp = _.gdpPercap) >> _.head()
```

```python
data = (
    data
    >> rename(life_expectancy = _.lifeExp, 
              gdp = _.gdpPercap)
)
```

## select() funksiyasÄ±

```python
data >> select(_.country, _.life_expectancy) >> _.head()
```

```python
data >> select(-_.country, -_.life_expectancy) >> _.head()
```

```python
data >> select(*data.columns[2:5]) >> _.head()
```

```python
cols_front = ['year', 'gdp']
cols_rest = [col for col in data.columns if col not in cols_front]
final_cols = cols_front + cols_rest

print(final_cols)

data >> select(*final_cols) >> _.head()
```

```python
cols = [col for col in data.columns if "_" in col]

data >> select(*cols) >> _.head()
```

```python
cols = [col for col in data.columns if col.startswith("co")]

data >> select(*cols) >> _.head()
```

```python
cols = [col for col in data.columns if col.endswith("p")]

data >> select(*cols) >> _.head()
```

## filter() funksiyasÄ±

```python
data >> filter(_.year > 2001) >> _.head()
```

```python
data >> filter(_.country == 'Albania') >> _.head()
```

```python
(
    data 
    >> filter(_.year > 2001,
              _.country == 'Albania')
)
```

```python
(
    data 
    >> filter(_.year.isin([2001, 2005, 2007]),
             ~_.continent.isin(["Europe", "Asia", "Africa"])) 
    >> _.head()
)
```

## mutate() funksiyasÄ±

```python
(
    data
    >> mutate(yeni_sÃ¼tun = _.life_expectancy / 100)
    >> _.head()
)
```

```python
(
    data
    >> mutate(yeni_sÃ¼tun = lambda x: round(x.life_expectancy / 100, 2))
    >> _.head()
)
```

## summarise() funksiyasÄ±

```python
(
    data
    >> summarize(max_life_expectancy = _.life_expectancy.max())
)
```

```python
(
    data
    >> summarize(
        max_life = _.life_expectancy.max(),
        mean_life = _.life_expectancy.mean(),
        min_life = _.life_expectancy.min()
    )
)
```

```python
(
    data 
    >> summarize(
        max = lambda x: round(x.life_expectancy.max(), 1),
        ortalama = lambda x: round(x.life_expectancy.mean(), 1),
        min = lambda x: round(x.life_expectancy.min(), 1)
    )
)
```

## group_by() funksiyasÄ±

```python
(
    data 
    >> group_by(_.country)
    >> summarize(max_life_expectancy = _.life_expectancy.max()) 
    >> _.head()
)
```

```python
(
    data 
    >> group_by(_.country) 
    >> summarize(
        max_life = _.life_expectancy.max(),
        mean_life = _.life_expectancy.mean(),
        min_life = _.life_expectancy.min()
    ) 
    >> _.head()
)
```

```python
(
    data
    >> group_by(_.country)
    >> summarize(max_life_ex = _.life_expectancy.max())
    >> arrange(-_.max_life_ex)
    >> _.head()
)
```

```python
(
    data
    >> group_by(_.continent, _.country)
    >> summarize(max_life_ex = _.life_expectancy.max())
    >> arrange(_.continent, -_.max_life_ex)
    >> _.head()
)
```

```python
(
    data
    >> summarize(
        sum_gdp = lambda x: x.gdp.sum(skipna=True),
        mean_gdp = lambda x: x.gdp.mean(skipna=True),
        med_gdp = lambda x: x.gdp.median(skipna=True)
    )
)
```

```python
(
    data
    >> group_by(_.continent)
    >> summarize(n = lambda x: x.shape[0])
)
```

```python
(
    data
    >> group_by(_.continent)
    >> summarize(n = lambda x: x.shape[0])
    >> arrange(-_.n)
)
```
