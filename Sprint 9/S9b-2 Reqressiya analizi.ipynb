```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
```

## Ardıcıl reqressiya

```python
stud = pd.read_csv("students.csv")
```

```python
fit1 = smf.ols('score ~ iq', data=stud).fit()
fit2 = smf.ols('score ~ iq + hours', data=stud).fit()
fit3 = smf.ols('score ~ iq + hours + gender', data=stud).fit()
```

```python
print(fit1.summary())
print(fit2.summary())
print(fit3.summary())
```

```python
# ANOVA ilə model müqayisəsi
anova_results = anova_lm(fit1, fit2, fit3)
print("\nANOVA Model Müqayisəsi:")
print(anova_results)
```

```python
# AIC və BIC ilə model müqayisəsi
comparison = pd.DataFrame({
    'Model': ['Model 1', 'Model 2', 'Model 3'],
    'AIC': [fit1.aic, fit2.aic, fit3.aic],
    'BIC': [fit1.bic, fit2.bic, fit3.bic]
})
print("\nAIC və BIC Müqayisəsi:")
print(comparison)
```

## "Binomial/Logistic" reqressiya

```python
mobi = pd.read_csv("mobilenet.csv")
```

```python
model = smf.logit('mobile ~ income + hours + C(where)', data=mobi).fit()
print(model.summary())
```

```python
mobi['pred_prob'] = model.predict(mobi)
mobi['pred_label'] = (mobi['pred_prob'] > 0.5).astype(int)
```

```python
confusion_matrix = pd.crosstab(mobi['mobile'], mobi['pred_label'], 
                               rownames=['Actual'], colnames=['Predicted'])
print("\nKonfuziya Matrisi:")
print(confusion_matrix)
```
